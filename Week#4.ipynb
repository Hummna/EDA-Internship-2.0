```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

menu_df = pd.read_csv('menu.csv')
```

```python
# Display statistical facts
print(menu_df.describe())

# Calculate maximum values for specified attributes
max_values = menu_df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 
                      'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 
                      'Calcium (% Daily Value)', 'Iron (% Daily Value)']].max()
print("\nMaximum values:")
print(max_values)
```

```python
# Calculate correlation matrix, excluding non-numeric columns
correlation_matrix = menu_df.drop(columns=['Category', 'Item', 'Serving Size']).corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Generate a list of unique categories in your data
categories = menu_df['Category'].unique()

# Draw boxplot with specified palette
sns.boxplot(x='Calories', y='Category', data=menu_df, palette='Set3')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Set plot title and labels
plt.title('Boxplot of Calories by Category')
plt.xlabel('Calories')
plt.ylabel('Category')

# Manually create legend with colors corresponding to categories
legend_labels = [plt.Rectangle((0,0),1,1,fc=sns.color_palette('Set3')[i], edgecolor = 'none') for i in range(len(categories))]
plt.legend(legend_labels, categories)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Draw strip plot for Calories by Category
sns.stripplot(x='Category', y='Calories', data=menu_df, jitter=True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set plot title and labels
plt.title('Strip Plot of Calories by Category')
plt.xlabel('Category')
plt.ylabel('Calories')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate average calories for each category
avg_calories = menu_df.groupby('Category')['Calories'].mean().sort_values()

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Plot horizontal bar graph
sns.barplot(x=avg_calories.values, y=avg_calories.index, palette='viridis')

# Set plot title and labels
plt.title('Average Calories by Category')
plt.xlabel('Average Calories')
plt.ylabel('Category')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Initialize an empty list to store selected items
selected_items = []

# Iterate through each category
categories = menu_df['Category'].unique()
for category in categories:
    # Filter DataFrame for items in the current category
    category_df = menu_df[menu_df['Category'] == category]
    # Select one item with the highest calorie count from the current category
    selected_item = category_df.loc[category_df['Calories'].idxmax()]
    # Add the selected item to the list
    selected_items.append(selected_item)

# Convert the list of selected items into a DataFrame
selected_df = pd.DataFrame(selected_items)

# Plot horizontal bar graph for selected items
sns.barplot(x='Calories', y='Item', data=selected_df, color='skyblue')

# Set plot title and labels
plt.title('Calories by One Item per Category')
plt.xlabel('Calories')
plt.ylabel('Item')

# Show the plot
plt.tight_layout()
plt.show()
```
