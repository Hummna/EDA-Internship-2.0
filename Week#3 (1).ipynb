```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
x = [1, 2, 3]
y = [2, 4, 1]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Graph!')

# to fix the sizes
plt.xlim(1.0, 3.0) 
plt.ylim(1.0, 4.0)  

plt.show()
```

```python
x = [1, 2, 3]
y1 = [2, 4, 1]
y2 = [4, 1, 3]
plt.xlim(1.0, 3.0) 
plt.ylim(1.0, 4.0) 

line1 = plt.plot(x, y1, label='Line 1', color='blue', linewidth=3)
line2 = plt.plot(x, y2, label='Line 2', color='red', linewidth=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two or more lines with legends, different widths and colors.')

# Creating custom legend 
custom_lines = [Line2D([0], [0], color='blue', lw=3),
                Line2D([0], [0], color='red', lw=5)]
plt.legend(custom_lines, ['Line1-width-3', 'Line2-width-5'])

plt.show()
```

```python
x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]

plt.plot(x, y, linestyle='dotted', marker='o', color='red', label='Line with Dotted Pattern', markerfacecolor='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(1, 8) 
plt.ylim(1, 8)
plt.title('Display Marker')
plt.legend()
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JS', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(languages, popularity)
plt.barh(languages, popularity, color='green')
plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('Popularity of Programming Languages Wordwild, Oct 2017 Compared to a year ago')
plt.show()
```

```python
data = {
    'a': [4, 2, 4, 2, 2],
    'b': [8, 2, 4, 6, 4],
    'c': [5, 3, 7, 4, 3],
    'd': [7, 2, 4, 8, 3],
    'e': [6, 6, 8, 6, 2]
}

df = pd.DataFrame(data)

# Assigning colors to each column
colors = ['blue', 'green', 'red', 'skyblue', 'purple']

df.plot(kind='bar', color=colors)
plt.show()
```

```python
data = {
    'country': ['United States', 'Great Britain', 'China', 'Russia', 'Germany'],
    'gold_medal': [46, 27, 26, 19, 17]
}

# Extracting data
countries = data['country']
medals = data['gold_medal']

# Distint United States wedge
explode = [0.1 if country == 'United States' else 0 for country in countries]

# Create pie chart with shadows
plt.figure(figsize=(8, 8))
plt.pie(medals, labels=countries, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title('Gold Medal Achievements of Top 5 Countries (2016 Summer Olympics)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
