```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv("births.csv")
```

```python
# Q.1: Add a new column “Decade”
data['Decade'] = (data['year'] // 10) * 10
```

```python
# Q.2: Descriptive statistics
descriptive_stats = data.describe()
```

```python
# Q.3: Check for missing values
missing_values = data.isnull().sum()
```

```python
# Q.4: Trend of male & female births every decade
trend_by_decade = data.groupby(['Decade', 'gender'])['births'].sum().unstack()
trend_by_decade.plot(kind='bar', stacked=True)
plt.title('Trend of Male & Female Births Every Decade')
plt.xlabel('Decade')
plt.ylabel('Total Births')
plt.show()
```

```python
# Q.5: Remove outliers
mean = data['births'].mean()
std_dev = data['births'].std()
data_no_outliers = data[np.abs(data['births'] - mean) <= 5 * std_dev]
```

```python
# Q.7: Group the data by month and day separately
births_by_month = data.groupby(['month'])['births'].sum()
births_by_day = data.groupby(['day'])['births'].sum()
```

```python
# Q.8: Plot average number of births by date of the year
births_by_date = data.groupby(['month', 'day'])['births'].mean()
births_by_date.plot(figsize=(12, 6))
plt.title('Average Number of Births by Date of the Year')
plt.xlabel('Date')
plt.ylabel('Average Number of Births')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
